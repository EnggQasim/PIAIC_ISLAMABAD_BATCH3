# Exception Handling

```python
print("Start.......")

try:
    print(6 / 0)
except ZeroDivisionError:  
    print("Ap Zero sy Divide nhi kr sakty!")
    

print("End......")    
```

```python
print("Start.......")

try:
    a = [2,3]
    print(a[7])
except IndexError:  
    print("Ap Zero sy Divide nhi kr sakty!")
    

print("End......") 
```

```python
print("Start.......")

try:
    open('abc.txt')
except FileNotFoundError:  
    print("Ap Zero sy Divide nhi kr sakty!")
    

print("End......")
```

```python
print("Start.......")

try:
    open('abc.txt')
    a = [2,3]
    print(a[7])
    print(6 / 0)
    
except (FileNotFoundError, ZeroDivisionError,IndexError):  
    print("ABC")
    

print("End......")
```
