```python
# A.I -> sum
# AIOT ->product
# CNCI -> cumsum
# else ->0

def custum_fn(x1, *l1):
    if x1=="A.I":
        return np.sum(l1)
    elif x1=="AIOT":
        return np.product(l1)
    elif x1=="CNC":
        return sum(np.cumsum(l1))
    else:
        return 0

  
```

```python
def custum_fn(x1, x2, x3, x4,x5, x6):
    if x1=="A.I":
        return x2+x3+x4+x5+x6
    elif x1=="AIOT":
        return x2*x3*x4*x5*x6
    elif x1=="CNC":
        return sum(np.cumsum([x2, x3, x4,x5, x6]))
    else:
        return 0
```

```python
df[['course','c1','c2','c3','c4','c5']].apply(lambda x:custum_fn(*x), axis=1)
```

```python
df["CheckValidation"] = df[['course','c1','c2','c3','c4','c5']].apply(lambda x:custum_fn(*x), axis=1)
df
```
