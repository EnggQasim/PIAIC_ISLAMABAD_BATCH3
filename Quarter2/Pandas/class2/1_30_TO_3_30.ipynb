```python
df2 = pd.DataFrame(np.random.randint(1,100, size=(5,5)), columns=["A","B","C","D","E"])
df2
```

# pd.concat

```python
df = pd.concat([df1,df2])
df
```

```python
df = pd.concat([df1,df2], axis=1)
df
```

```python
df['total'] = df.A + df.B + df.C + df.D + df.E
df['percentage'] = (df.A + df.B + df.C + df.D + df.E)*100/500

def grade(p):
    if p>= 80: return "A+"
    elif p>=70: return "A" 
    elif p>=60: return "B"
    elif p>=50: return "C"
    else: return "Fail"

df['Grade'] = df.percentage.apply(grade)
df
```

# How to apply function on multiple columns in pandas

```python
df1 = df.iloc[:,:-3]
df1
```

```python
def grade1(x1,x2,x3,x4,x5):
    total = x1+x2+x3+x4+x5
    p = total/500*100
    
    if p>= 80: return "A+"
    elif p>=70: return "A" 
    elif p>=60: return "B"
    elif p>=50: return "C"
    else: return "Fail"
    
df1['Grade']=df1[["A","B","C","D","E"]].apply(lambda x:grade1(*x), axis=1)
df1
```
