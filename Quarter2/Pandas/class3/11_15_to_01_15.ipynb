```python
customers = pd.DataFrame({"Customers NIC": [14,6,19],"Customer Name":["A","B","C"]})


transactions = pd.DataFrame({"CNIC": np.random.choice([14,6,19,21], size=(20)),
                   "TransactionID":np.random.randint(1,2, size=(20)),
                  "Transaction Date":pd.date_range('1990-01-01', periods=20, freq='Y'),
                   "Description":np.random.choice(["M","F"], size=(20)),
                            "Actual Amount":np.random.randint(1,10000, size=(20))})


transactions.columns = ['CNIC', 'TransactionID', 'Transaction Date', 'Gender','Actual Amount']
transactions
```

# Group By

```python
#dataframeName.groupby(by="").agg([count,min,max])
#dataframeName.groupby(by=[col1,col2,..]).agg([count,min,max])

c = transactions.groupby("Gender").agg(['count'])
c
```

```python
%timeit c = transactions.groupby("Gender")[["CNIC"]].agg(['count'])
%timeit c
```

```python
c = transactions.groupby([transactions['Transaction Date'].dt.year  ,  "Gender"])[["CNIC"]].agg(['count'])
c
```

```python
c = transactions.groupby(["Gender","CNIC"])[["Actual Amount"]].agg(['count',"sum",'min','max'])
c
```
